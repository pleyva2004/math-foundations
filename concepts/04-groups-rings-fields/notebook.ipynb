# Groups, Rings, Fields

Concept 04 of the math-foundations learning map.

This notebook walks through the four group axioms on the cyclic
group $\mathbb{Z}/n\mathbb{Z}$, then shows that $\mathbb{Z}/p\mathbb{Z}$
is a field for prime $p$ but $\mathbb{Z}/n\mathbb{Z}$ is not a field
for composite $n$.

## Group axioms

A group $(G, \cdot)$ satisfies:

1. **Closure**: $a \cdot b \in G$ for all $a, b \in G$.
2. **Associativity**: $(a \cdot b) \cdot c = a \cdot (b \cdot c)$.
3. **Identity**: $\exists\, e \in G$ with $e \cdot a = a \cdot e = a$.
4. **Inverses**: $\forall a \in G,\ \exists\, a^{-1}$ with $a \cdot a^{-1} = e$.

Below we verify these for $(\mathbb{Z}/n\mathbb{Z}, +)$.

```python
class Zmod:
    def __init__(self, n):
        self.n = n
        self.elements = list(range(n))
    def add(self, a, b): return (a + b) % self.n
    def mul(self, a, b): return (a * b) % self.n
    def neg(self, a):    return (-a) % self.n

def verify_group_axioms(G):
    e = G.elements
    closure = all(G.add(a, b) in e for a in e for b in e)
    assoc   = all(G.add(G.add(a, b), c) == G.add(a, G.add(b, c))
                  for a in e for b in e for c in e)
    iden    = all(G.add(0, a) == a and G.add(a, 0) == a for a in e)
    inv     = all(G.add(a, G.neg(a)) == 0 for a in e)
    return closure, assoc, iden, inv

G5 = Zmod(5)
print('Axioms for Z/5Z under + :', verify_group_axioms(G5))
```

## From group to field

A **ring** $(R, +, \cdot)$ is an abelian group under $+$ together with an
associative multiplication that distributes over $+$. A **field** is a
commutative ring in which every nonzero element has a multiplicative
inverse.

Claim: $\mathbb{Z}/5\mathbb{Z}$ is a field; $\mathbb{Z}/6\mathbb{Z}$ is not.
We test this by brute-force searching for inverses.

```python
def mul_inverse(G, a):
    for b in G.elements:
        if G.mul(a, b) == 1:
            return b
    return None

for n in (5, 6):
    G = Zmod(n)
    inverses = {a: mul_inverse(G, a) for a in G.elements if a != 0}
    missing  = [a for a, b in inverses.items() if b is None]
    print(f'Z/{n}Z inverses of nonzero elements: {inverses}')
    print(f'  field? {len(missing) == 0}; missing = {missing}')
    print()
```
