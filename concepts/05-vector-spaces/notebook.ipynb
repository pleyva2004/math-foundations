# Vector Spaces (concept 05)

A **vector space** over a field $\mathbb{F}$ is a set $V$ closed under vector addition and scalar multiplication, satisfying the eight axioms (associativity, commutativity, identities, inverses, distributivity, compatibility). This notebook explores bases, linear independence, dimension, and change of basis in $\mathbb{R}^3$.

Prereq: [04-groups-rings-fields](../04-groups-rings-fields/README.md).

## Axioms

For all $u, v, w \in V$ and $\alpha, \beta \in \mathbb{F}$:

1. $(u + v) + w = u + (v + w)$
2. $u + v = v + u$
3. There exists $0 \in V$ with $v + 0 = v$
4. For each $v$ there exists $-v$ with $v + (-v) = 0$
5. $\alpha(u + v) = \alpha u + \alpha v$
6. $(\alpha + \beta) v = \alpha v + \beta v$
7. $(\alpha \beta) v = \alpha(\beta v)$
8. $1 \cdot v = v$

**Basis**: a linearly independent spanning set.  
**Dimension**: cardinality of any basis (well-defined by the Steinitz exchange lemma).

```python
import numpy as np

# Standard basis of R^3
E = np.eye(3)
print('Standard basis (rows):')
print(E)
print('rank =', np.linalg.matrix_rank(E), '(equals dim R^3 = 3)')

# A non-standard basis
B = np.array([[1.0, 1.0, 0.0],
              [0.0, 1.0, 1.0],
              [1.0, 0.0, 1.0]])
print('\nNon-standard basis (rows):')
print(B)
print('det =', np.linalg.det(B), '(nonzero => independent)')
print('rank =', np.linalg.matrix_rank(B))
```

## Worked example: change of basis

Let $B = \{b_1, b_2, b_3\}$ be a basis of $\mathbb{R}^3$. Every $v \in \mathbb{R}^3$ has unique coordinates $\alpha \in \mathbb{R}^3$ with
$$v = \sum_{i=1}^{3} \alpha_i\, b_i.$$
Writing the $b_i$ as columns of a matrix $M$, this becomes $v = M \alpha$, so $\alpha = M^{-1} v$.

Concretely: $v = (2, 5, 7)^\top$ in the basis $\{(1,1,0), (0,1,1), (1,0,1)\}$ has coordinates we now compute.

```python
# Express v in the non-standard basis
B = np.array([[1.0, 1.0, 0.0],
              [0.0, 1.0, 1.0],
              [1.0, 0.0, 1.0]])
v = np.array([2.0, 5.0, 7.0])

# Solve B^T alpha = v  (rows of B are basis vectors)
alpha = np.linalg.solve(B.T, v)
print('coordinates alpha =', alpha)

# Reconstruct
recon = alpha @ B
print('reconstruction =', recon)
print('matches v?', np.allclose(recon, v))

# Dimension theorem: both bases have size 3
print('\n|standard basis| =', 3)
print('|non-standard basis| =', len(B))
```
