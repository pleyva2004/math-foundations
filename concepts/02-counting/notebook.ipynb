# Counting and Combinatorics

Concept **02-counting** of the math-foundations learning map. This notebook
walks through permutations, combinations, the binomial theorem, and the
pigeonhole principle with executable demonstrations.


## Definitions

- **Permutation:** ordered arrangement. $P(n,k) = n! / (n-k)!$.
- **Combination:** unordered subset. $\binom{n}{k} = n! / (k!\,(n-k)!)$.
- **Multinomial coefficient:** $\binom{n}{k_1,\ldots,k_r} = n! / \prod_i k_i!$.
- **Pigeonhole principle:** if $m$ items are placed in $n < m$ boxes, some box has $\geq 2$ items.
- **Inclusion-exclusion:** $|A \cup B| = |A| + |B| - |A \cap B|$, generalized via alternating signs.


```python
from itertools import permutations, combinations
from math import comb, perm

S = ('a', 'b', 'c', 'd', 'e')
n, k = len(S), 3

perms = list(permutations(S, k))
combs = list(combinations(S, k))

print(f'enumerated {len(perms)} permutations vs P({n},{k}) = {perm(n, k)}')
print(f'enumerated {len(combs)} combinations vs C({n},{k}) = {comb(n, k)}')
print('first few perms:', perms[:3])
print('first few combs:', combs[:3])
```

## Binomial theorem

$(1+x)^n = \sum_{k=0}^n \binom{n}{k} x^k$. We verify numerically at $n=6, x=2$.


```python
from math import comb

n, x = 6, 2
lhs = (1 + x) ** n
rhs = sum(comb(n, k) * x ** k for k in range(n + 1))
print(f'(1+{x})^{n} = {lhs}')
print(f'sum_k C({n},k) {x}^k = {rhs}')
assert lhs == rhs
```

## Pigeonhole principle

Place 10 items into 7 bins. By pigeonhole some bin must contain at least
$\lceil 10/7 \rceil = 2$ items.


```python
from collections import Counter

items = list(range(10))
bins = 7
assignment = {i: i % bins for i in items}
counts = Counter(assignment.values())
for b in sorted(counts):
    print(f'bin {b}: {counts[b]} item(s)')
print('fullest bin holds', max(counts.values()), 'items')
assert max(counts.values()) >= 2
```
