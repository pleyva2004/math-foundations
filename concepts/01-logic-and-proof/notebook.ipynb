## 1. Propositional logic — definitions

A **proposition** is a sentence with a definite truth value in $\{\text{True}, \text{False}\}$.
The binary connectives we use throughout this notebook are:

| Connective | Symbol | Python |
|---|---|---|
| Conjunction | $p \land q$ | `p and q` |
| Disjunction | $p \lor q$ | `p or q` |
| Implication | $p \Rightarrow q$ | `(not p) or q` |
| Biconditional | $p \Leftrightarrow q$ | `p == q` |
| Negation | $\neg p$ | `not p` |

Implication is `False` only when $p$ is True and $q$ is False — a *false hypothesis implies anything*.


```python
from itertools import product

def NEG(p): return not p
def AND(p, q): return p and q
def OR(p, q):  return p or q
def IMP(p, q): return (not p) or q
def IFF(p, q): return p == q

def truth_table(name, op):
    print(f"{name}")
    print("  p     q     result")
    for p, q in product([False, True], repeat=2):
        print(f"  {str(p):5} {str(q):5} {op(p, q)}")
    print()

for nm, op in [("AND", AND), ("OR", OR), ("IMP (p => q)", IMP), ("IFF", IFF)]:
    truth_table(nm, op)
```

## 2. Tautologies

A compound proposition is a **tautology** if it evaluates to True under every truth assignment of its
atomic variables. Classic tautologies we will verify by brute force:

- *Law of excluded middle:* $\;p \lor \neg p$
- *Contrapositive equivalence:* $\;(p \Rightarrow q) \Leftrightarrow (\neg q \Rightarrow \neg p)$
- *De Morgan:* $\;\neg(p \land q) \Leftrightarrow (\neg p) \lor (\neg q)$


```python
def verify_tautology(name, predicate, n_vars):
    for assignment in product([False, True], repeat=n_vars):
        if not predicate(*assignment):
            print(f"FAIL  {name}  counterexample={assignment}")
            return False
    print(f"OK    {name}  (checked {2**n_vars} assignments)")
    return True

verify_tautology("p OR NOT p", lambda p: OR(p, NEG(p)), 1)
verify_tautology("(p => q) <=> (NOT q => NOT p)",
                 lambda p, q: IFF(IMP(p, q), IMP(NEG(q), NEG(p))), 2)
verify_tautology("NOT(p AND q) <=> (NOT p) OR (NOT q)",
                 lambda p, q: IFF(NEG(AND(p, q)), OR(NEG(p), NEG(q))), 2)
```

## 3. Mathematical induction — the running example

**Theorem (Principle of Mathematical Induction).** *Let $P(n)$ be a predicate on $\mathbb{N}$.
If $P(1)$ is true and $P(k) \Rightarrow P(k+1)$ for every $k \in \mathbb{N}$, then $P(n)$ holds for every $n \in \mathbb{N}$.*

**Claim.** For every $n \geq 1$,
$$
\sum_{k=1}^{n} k \;=\; \frac{n(n+1)}{2}.
$$

**Proof (sketch; full version in `lesson.tex`).**
*Base.* $n=1$: $\;1 = 1\cdot 2 / 2$. ✓
*Step.* Assume $\sum_{k=1}^{m} k = m(m+1)/2$. Then
$\sum_{k=1}^{m+1} k = m(m+1)/2 + (m+1) = (m+1)(m+2)/2$. ✓

A proof is *not* a numerical check — but numerical agreement on the first 20 cases is a useful
sanity test before you commit to writing out the induction.


```python
def closed_form(n): return n * (n + 1) // 2

print("  n |  explicit  |  closed    | equal?")
print(" ---+------------+------------+--------")
all_ok = True
for n in range(1, 21):
    explicit = sum(range(1, n + 1))
    closed = closed_form(n)
    ok = (explicit == closed)
    all_ok = all_ok and ok
    print(f"  {n:2d}|  {explicit:8d}  |  {closed:8d}  | {ok}")

print()
print(f"Identity verified for n = 1..20: {all_ok}")
```
