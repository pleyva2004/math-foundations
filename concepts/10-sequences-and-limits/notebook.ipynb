A sequence in a set $X$ is a function $x : \mathbb{N} \to X$, written $(x_n)$. We study what it means for the sequence to *converge* to a point $L \in X$, and the closely related *Cauchy* property, which lets us recognise convergent sequences without naming the limit.

This notebook is self-contained (stdlib only) and accompanies `README.md`, `lesson.tex`, and `code.py` in the same folder.

## The $\epsilon$-$N$ definition of convergence

Let $(X, d)$ be a metric space. A sequence $(x_n)$ in $X$ **converges** to $L \in X$ if
$$
\forall \epsilon > 0,\ \exists N \in \mathbb{N},\ \forall n \ge N,\ d(x_n, L) < \epsilon.
$$
Informally: however small a tolerance $\epsilon$ you name, *eventually* (from some index $N$ onward) every term lies within $\epsilon$ of $L$.

We will witness this concretely with $x_n = 1/n$, which converges to $0$ in $\mathbb{R}$: for each $\epsilon$, we hunt for the smallest $N$ that works.

```python
import math

def smallest_N_for_one_over_n(eps, n_max=200000):
    for N in range(1, n_max + 1):
        if 1.0 / N < eps:
            return N
    raise ValueError('no N found')

print(f"{'epsilon':>12} {'smallest N':>14} {'1/N':>16}")
for eps in (1e-2, 1e-3, 1e-4):
    N = smallest_N_for_one_over_n(eps)
    print(f'{eps:>12.6f} {N:>14d} {1.0/N:>16.8f}')
```

## Cauchy sequences and completeness

A sequence $(x_n)$ is **Cauchy** if
$$
\forall \epsilon > 0,\ \exists N \in \mathbb{N},\ \forall m, n \ge N,\ d(x_m, x_n) < \epsilon.
$$
The Cauchy condition is intrinsic: it talks only about how the terms relate to *each other*, not to any external limit.

**Key theorem.** Every convergent sequence is Cauchy (use the triangle inequality and split $\epsilon$ as $\epsilon/2 + \epsilon/2$).

**Completeness** of a metric space says the converse: every Cauchy sequence converges to a point of the space. The real line $\mathbb{R}$ is complete; $\mathbb{Q}$ is not.

Below we monitor the Basel partial sums $x_n = \sum_{k=1}^n 1/k^2$ (true limit $\pi^2/6$) and verify that $|x_m - x_n|$ shrinks as $m, n \to \infty$ -- the numerical signature of the Cauchy property.

```python
def basel_partial(n):
    s = 0.0
    for k in range(1, n + 1):
        s += 1.0 / (k * k)
    return s

limit = math.pi ** 2 / 6.0
print(f'True limit pi^2/6 = {limit:.10f}')
print()
print(f"{'m':>8} {'n':>8} {'|x_m - x_n|':>20}")
for m, n in [(100, 50), (1000, 500), (5000, 2500), (10000, 5000)]:
    diff = abs(basel_partial(m) - basel_partial(n))
    print(f'{m:>8d} {n:>8d} {diff:>20.3e}')
```
