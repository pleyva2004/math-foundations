# Eigenvalues and Eigenvectors

**Node 07** of the math-foundations learning map.

An *eigenvector* of a linear map $A$ is a non-zero direction that $A$ merely scales; the scaling factor is the *eigenvalue*.
Eigendecomposition underlies PCA, spectral clustering, the analysis of LoRA-style low-rank deltas, and the curvature analysis of optimisers.

## Definitions

Let $A \in \mathbb{R}^{n\times n}$.

- $\lambda \in \mathbb{R}$ (or $\mathbb{C}$) is an **eigenvalue** of $A$ if there exists $v \ne 0$ with $A v = \lambda v$.
- The non-zero $v$ is called an **eigenvector**.
- The **characteristic polynomial** is $p_A(\lambda) = \det(A - \lambda I)$; its roots are exactly the eigenvalues.
- $A$ is **diagonalisable** iff $\mathbb{R}^n$ admits a basis of eigenvectors of $A$.

```python
# Pure-stdlib eigendecomposition by Jacobi rotations (symmetric A).
# (See code.py for the full version; this is the same algorithm inline.)
from code import jacobi_eigh, matmul

A = [[4.0, 1.0, 2.0], [1.0, 3.0, 0.0], [2.0, 0.0, 5.0]]
eigvals, Q = jacobi_eigh(A)
print('Eigenvalues:', eigvals)
print('First eigenvector (column 0 of Q):', [Q[i][0] for i in range(3)])
```

## The spectral theorem

If $A \in \mathbb{R}^{n\times n}$ is symmetric ($A = A^\top$), then:

1. All eigenvalues of $A$ are real.
2. There exists an orthonormal basis of eigenvectors.
3. Equivalently, $A = Q \Lambda Q^\top$ with $Q$ orthogonal and $\Lambda$ diagonal.

This is the engine behind PCA: covariance matrices are symmetric PSD, so they have an orthonormal eigenbasis whose eigenvalues are the principal variances.

```python
import math
from code import jacobi_eigh, matmul

A = [[4.0, 1.0, 2.0], [1.0, 3.0, 0.0], [2.0, 0.0, 5.0]]
eigvals, Q = jacobi_eigh(A)
Qt = [[Q[j][i] for j in range(3)] for i in range(3)]
L = [[eigvals[i] if i == j else 0.0 for j in range(3)] for i in range(3)]
A_rec = matmul(matmul(Q, L), Qt)
err = math.sqrt(sum((A[i][j] - A_rec[i][j])**2 for i in range(3) for j in range(3)))
print('||A - Q Lambda Q^T||_F =', err)

# Orthogonality check.
QtQ = matmul(Qt, Q)
ortho = math.sqrt(sum((QtQ[i][j] - float(i==j))**2 for i in range(3) for j in range(3)))
print('||Q^T Q - I||_F =', ortho)
```
