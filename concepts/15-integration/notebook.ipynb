## 1. Riemann sum, formally

For a partition $P = \{x_0, \dots, x_n\}$ of $[a, b]$ and sample points
$t_i \in [x_{i-1}, x_i]$, the **Riemann sum** is
$$ S(f, P, t) = \sum_{i=1}^{n} f(t_i)\, \Delta x_i, \quad \Delta x_i = x_i - x_{i-1}. $$

Three common choices for $t_i$:
- **Left rule:** $t_i = x_{i-1}$.
- **Right rule:** $t_i = x_i$.
- **Midpoint rule:** $t_i = (x_{i-1} + x_i)/2$.

If $f$ is continuous on $[a, b]$, all three converge to the same limit $\int_a^b f$
as $\|P\| \to 0$.

```python
def f(x):
    return x * x

def riemann_sum(g, a, b, n, rule='midpoint'):
    h = (b - a) / n
    total = 0.0
    for i in range(n):
        if rule == 'left':
            x = a + i * h
        elif rule == 'right':
            x = a + (i + 1) * h
        else:  # midpoint
            x = a + (i + 0.5) * h
        total += g(x) * h
    return total

exact = 1.0 / 3.0
for n in (10, 100, 1000):
    L = riemann_sum(f, 0, 1, n, 'left')
    R = riemann_sum(f, 0, 1, n, 'right')
    M = riemann_sum(f, 0, 1, n, 'midpoint')
    print(f'n={n:5d}  L={L:.8f}  R={R:.8f}  M={M:.8f}  |M-1/3|={abs(M-exact):.2e}')
print(f'exact = {exact}')
```

## 2. Fundamental Theorem of Calculus

**FTC2.** If $f$ is continuous on $[a,b]$ and $G' = f$, then
$$ \int_a^b f(x)\, dx = G(b) - G(a). $$

For $f(x) = x^2$ an antiderivative is $G(x) = x^3 / 3$, so
$\int_0^1 x^2\, dx = G(1) - G(0) = 1/3 - 0 = 1/3$.
We verify this exactly (no limit needed).

```python
def G(x):
    return x**3 / 3

ftc_value = G(1) - G(0)
print('FTC2:  G(1) - G(0) =', ftc_value)
print('exact 1/3        =', 1/3)
print('match?           =', ftc_value == 1/3)
```
