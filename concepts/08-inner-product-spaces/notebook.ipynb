## 1. Definition of an inner product

An inner product on a real vector space $V$ is a map
$\langle \cdot, \cdot \rangle : V \times V \to \mathbb{R}$ that is

- **symmetric:** $\langle u, v \rangle = \langle v, u \rangle$,
- **linear in the first argument:** $\langle \alpha u + w, v \rangle = \alpha \langle u, v \rangle + \langle w, v \rangle$,
- **positive definite:** $\langle u, u \rangle \ge 0$ with equality iff $u = 0$.

The induced norm is $\|v\| = \sqrt{\langle v, v \rangle}$.
For $\mathbb{R}^n$ the canonical inner product is the dot product
$\langle u, v \rangle = \sum_{i=1}^n u_i v_i$.

```python
import numpy as np

def inner(u, v):
    return float(np.dot(u, v))

def norm(u):
    return float(np.sqrt(inner(u, u)))

# Cauchy-Schwarz numerical check on 1000 random pairs in R^5.
rng = np.random.default_rng(seed=42)
violations = 0
ratios = []
for _ in range(1000):
    a = rng.standard_normal(5)
    b = rng.standard_normal(5)
    lhs = abs(inner(a, b))
    rhs = norm(a) * norm(b)
    if lhs > rhs + 1e-9:
        violations += 1
    if rhs > 1e-12:
        ratios.append(lhs / rhs)
print('Cauchy-Schwarz violations:', violations, '/ 1000')
print('Max ratio observed       :', max(ratios))
```

## 3. Gram-Schmidt orthonormalisation

Given linearly independent $v_1, \dots, v_k$, build an orthonormal
basis $\{e_1, \dots, e_k\}$ for their span by

$$u_j = v_j - \sum_{i=1}^{j-1} \langle v_j, e_i \rangle\, e_i,
  \qquad e_j = u_j / \|u_j\|.$$

Below we apply Gram-Schmidt to three vectors in $\mathbb{R}^3$ and
verify the result is orthonormal by checking $Q Q^\top = I$.

```python
def gram_schmidt(vectors):
    basis = []
    for v in vectors:
        w = v.astype(float).copy()
        for e in basis:
            w = w - inner(w, e) * e
        basis.append(w / norm(w))
    return basis

v1 = np.array([1.0, 1.0, 0.0])
v2 = np.array([1.0, 0.0, 1.0])
v3 = np.array([0.0, 1.0, 1.0])
basis = gram_schmidt([v1, v2, v3])
Q = np.stack(basis, axis=0)
print('Orthonormal basis (rows):')
print(np.round(Q, 4))
print('Q Q^T (should be identity):')
print(np.round(Q @ Q.T, 6))
```
