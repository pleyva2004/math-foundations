## Definitions (recap)

A **set** $A$ is an unordered collection of distinct elements. Its **power set** is
$$
\mathcal{P}(A) = \{ S : S \subseteq A \}, \qquad |\mathcal{P}(A)| = 2^{|A|} \text{ when } A \text{ is finite.}
$$

A **function** $f : A \to B$ is a subset $f \subseteq A \times B$ such that for every $a \in A$ there is a unique $b \in B$ with $(a,b) \in f$. We classify $f$ as:

- **injective**: $f(a_1) = f(a_2) \implies a_1 = a_2$
- **surjective**: $f(A) = B$
- **bijective**: both of the above

Cantor's theorem says $|\mathcal{P}(A)| > |A|$ for *every* set $A$ -- there is no surjection $A \to \mathcal{P}(A)$.


```python
from itertools import chain, combinations

def power_set(a):
    elems = list(a)
    return [frozenset(c) for c in chain.from_iterable(
        combinations(elems, r) for r in range(len(elems) + 1))]

def is_injective(f, domain):
    seen = {}
    for x in domain:
        y = f(x)
        if y in seen and seen[y] != x:
            return False
        seen[y] = x
    return True

def is_surjective(f, domain, codomain):
    return set(codomain).issubset({f(x) for x in domain})

def fmt_set(s):
    return '{' + ', '.join(str(x) for x in sorted(s)) + '}' if s else '{}'

A = {0, 1, 2, 3}
pA = power_set(A)
print(f'|A| = {len(A)}, |P(A)| = {len(pA)}, 2^|A| = {2**len(A)}')
assert len(pA) == 2 ** len(A)
```

## Worked example

Define $f : \{0,1,2,3\} \to \{0,1\}$ by $f(x) = x \bmod 2$.

Then $f(0) = 0$, $f(1) = 1$, $f(2) = 0$, $f(3) = 1$. The image is $\{0,1\}$, so $f$ is surjective. But $f(0) = f(2)$ with $0 \ne 2$, so $f$ is not injective. The next cell verifies this programmatically.


```python
f = lambda x: x % 2
B = {0, 1}
print('f: A -> B, f(x) = x % 2')
print('  injective? ', is_injective(f, A))
print('  surjective?', is_surjective(f, A, B))

# Bijection check: g(x) = (x + 1) mod 4 permutes A.
g = lambda x: (x + 1) % 4
print('\ng: A -> A, g(x) = (x + 1) mod 4')
print('  injective? ', is_injective(g, A))
print('  surjective?', is_surjective(g, A, A))

# Characteristic-function bijection P(A) <-> {0,1}^A.
print('\nCharacteristic functions (subset <-> bitstring):')
for s in pA[:6]:
    bits = ''.join('1' if x in s else '0' for x in sorted(A))
    label = fmt_set(s)
    print(f'  {label:>14}  <->  {bits}')
```
