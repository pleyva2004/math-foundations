# Sample Spaces — Interactive Notebook

Concept **17-sample-spaces** of the math-foundations learning map.

A *sample space* $\Omega$ is the set of all possible outcomes of an experiment. Every probability claim is implicitly relative to some $\Omega$.


## 1. Building finite sample spaces

We start with the canonical examples: a coin, a die, and two dice (a product space).


```python
from itertools import product, combinations
import random

coin = ('H', 'T')
die = (1, 2, 3, 4, 5, 6)
two_dice = tuple(product(die, repeat=2))

print('|coin|     =', len(coin))
print('|die|      =', len(die))
print('|two_dice| =', len(two_dice))
two_dice[:6]
```

## 2. The event family is the power set

For a finite sample space the family of events is the power set $2^\Omega$. We enumerate it explicitly and verify that $|2^\Omega| = 2^{|\Omega|}$.


```python
def power_set(omega):
    items = list(omega)
    return [frozenset(c) for r in range(len(items)+1) for c in combinations(items, r)]

for name, omega in [('coin', coin), ('die', die)]:
    ev = power_set(omega)
    print(f'{name}: |Omega|={len(omega)}, |2^Omega|={len(ev)} = 2^{len(omega)}')
    assert len(ev) == 2 ** len(omega)
```

## 3. Events as subsets

An event is just a subset of $\Omega$. Let us pick out the event "sum equals 7" in the two-dice sample space.


```python
sum7 = [p for p in two_dice if sum(p) == 7]
print('Event {sum = 7}:', sum7)
print('Cardinality:', len(sum7))
print('Probability under uniform measure:', len(sum7) / len(two_dice))
```

## 4. Empirical frequencies approach uniform probability

If outcomes are equally likely, the empirical frequency of any single outcome should approach $1/|\Omega|$ as the number of trials grows (law of large numbers).


```python
def simulate(omega, trials, seed=0):
    rng = random.Random(seed)
    counts = {w: 0 for w in omega}
    for _ in range(trials):
        counts[rng.choice(list(omega))] += 1
    return {w: c/trials for w, c in counts.items()}

for trials in (100, 10_000, 1_000_000):
    freqs = simulate(die, trials, seed=42)
    max_dev = max(abs(p - 1/6) for p in freqs.values())
    print(f'{trials:>8d} trials  max |freq - 1/6| = {max_dev:.5f}')
```

## 5. Why this matters for ML

In reinforcement learning, the sample space of trajectories is
$$\Omega = (\mathcal{S} \times \mathcal{A})^{\mathbb{N}}.$$
Below we build the *one-step* RL sample space $\mathcal{S} \times \mathcal{A}$ for a toy problem with 3 states and 2 actions.


```python
S = ('s0', 's1', 's2')
A = ('left', 'right')
Omega_rl = tuple(product(S, A))
print('|S x A| =', len(Omega_rl))
print(Omega_rl)
```
