# Norms and Metrics (concept 09)

Generalised notions of distance --- the foundation of analysis on vector spaces.

This notebook walks through:

1. The formal definitions of **norm** and **metric**.
2. Three classical norms on $\mathbb{R}^n$: $\ell^1$, $\ell^2$, $\ell^\infty$.
3. A numerical sanity-check of the triangle inequality.
4. The **discrete metric**, which is a metric but not norm-induced.
5. Exercises for self-study.


## 1. Definitions

A **norm** $\|\cdot\| : V \to \mathbb{R}$ on a vector space $V$ satisfies:

- $\|x\| \ge 0$, with equality iff $x = 0$ (positive definiteness);
- $\|\alpha x\| = |\alpha| \, \|x\|$ (absolute homogeneity);
- $\|x + y\| \le \|x\| + \|y\|$ (triangle inequality).

A **metric** $d : X \times X \to \mathbb{R}$ on a set $X$ satisfies:

- $d(x,y) \ge 0$, with equality iff $x = y$;
- $d(x,y) = d(y,x)$ (symmetry);
- $d(x,z) \le d(x,y) + d(y,z)$ (triangle inequality).

**Induced metric.** If $\|\cdot\|$ is a norm on $V$, then
$d(x,y) := \|x - y\|$ is a metric. (Proof in `lesson.tex`.)


```python
import math, random

def l1(x):   return sum(abs(c) for c in x)
def l2(x):   return math.sqrt(sum(c*c for c in x))
def linf(x): return max(abs(c) for c in x) if x else 0.0

v = [3.0, -4.0, 12.0]
print(f"v          = {v}")
print(f"||v||_1    = {l1(v)}")
print(f"||v||_2    = {l2(v)}")
print(f"||v||_inf  = {linf(v)}")
print(f"||v||_inf <= ||v||_2 <= ||v||_1 : {linf(v) <= l2(v) <= l1(v)}")
```

## 2. The triangle inequality, numerically

For any norm we must have $\|x + y\| \le \|x\| + \|y\|$.
Below we sample 200 random pairs in $\mathbb{R}^5$ and check this
empirically for all three norms. The reported *slack* is
$\|x\| + \|y\| - \|x+y\|$, which is always $\ge 0$.


```python
def vadd(x, y): return [a + b for a, b in zip(x, y)]

def check_triangle(norm, trials=200, dim=5, seed=0):
    rng = random.Random(seed)
    worst = math.inf
    for _ in range(trials):
        x = [rng.uniform(-10, 10) for _ in range(dim)]
        y = [rng.uniform(-10, 10) for _ in range(dim)]
        lhs = norm(vadd(x, y))
        rhs = norm(x) + norm(y)
        assert lhs <= rhs + 1e-9
        worst = min(worst, rhs - lhs)
    return worst

for name, fn in [("l1", l1), ("l2", l2), ("linf", linf)]:
    print(f"{name:5s} passed; smallest slack = {check_triangle(fn):.4f}")
```

## 3. The discrete metric

Define $d(x,y) = 0$ if $x = y$, else $d(x,y) = 1$. This is a metric on
any set, but on a (nonzero) vector space it cannot come from a norm,
because absolute homogeneity would force $\|2x\| = 2\|x\|$, while
$d(2x,0) = 1 = d(x,0)$ for any $x \ne 0$.


```python
def discrete(x, y):
    return 0 if x == y else 1

x = (1.0, 0.0, 0.0)
zero = (0.0, 0.0, 0.0)
two_x = (2.0, 0.0, 0.0)

n_x  = discrete(x, zero)
n_2x = discrete(two_x, zero)
print(f"d(x, 0)  = {n_x}")
print(f"d(2x, 0) = {n_2x}")
print(f"homogeneity needs {n_2x} == 2 * {n_x} = {2*n_x}: {n_2x == 2*n_x}")
print("=> the discrete metric is NOT induced by any norm.")
```
