# Series and Convergence

Concept **16-series-and-convergence** of the math-foundations learning map.

A *series* is the limit of its partial sums. We explore convergence, absolute
vs conditional convergence, and Taylor expansions — the bridge from analysis
to the smooth functions that appear everywhere in ML (activations, the matrix
exponential, KL near identity, softmax in concentrated regimes).


## 1. Series definitions

Given a sequence $(a_n)_{n \ge 0}$:

- **Partial sum:** $s_n = \sum_{k=0}^{n} a_k$.
- **Convergence:** $\sum a_n = S$ iff $s_n \to S$.
- **Absolute convergence:** $\sum |a_n|$ converges. Implies convergence.
- **Conditional convergence:** $\sum a_n$ converges but $\sum |a_n|$ does not.
- **Geometric series:** $\sum_{n=0}^{\infty} r^n = 1/(1-r)$ for $|r| < 1$.
- **Power series:** $\sum c_n (x - x_0)^n$ with radius of convergence
  $1/R = \limsup |c_n|^{1/n}$ (Cauchy–Hadamard).

The **ratio test** says: if $L = \lim |a_{n+1}/a_n|$ exists with $L < 1$, then
$\sum a_n$ converges absolutely. The proof compares $|a_n|$ to a geometric
series with ratio $r \in (L, 1)$.


## 2. Partial sums in code

We watch three canonical series:

1. $\sum 1/n^2 \to \pi^2/6$ (Basel — converges absolutely).
2. $\sum 1/n$ — harmonic, **diverges** like $\ln n$.
3. $\sum (-1)^n / n = -\ln 2$ — **conditionally** convergent.


```python
import math

def partial_sums(term, n_terms):
    s = 0.0
    for k in range(1, n_terms + 1):
        s += term(k)
        yield k, s

checkpoints = {10, 100, 1_000, 10_000, 100_000}

print("Basel sum 1/n^2 -> pi^2/6 =", math.pi**2 / 6)
for k, s in partial_sums(lambda n: 1.0 / n**2, 100_000):
    if k in checkpoints:
        print(f"  n={k:>7d}  s_n={s:.10f}  err={abs(s - math.pi**2/6):.2e}")

print("\nHarmonic sum 1/n diverges; compare to ln(n)")
for k, s in partial_sums(lambda n: 1.0 / n, 100_000):
    if k in checkpoints:
        print(f"  n={k:>7d}  s_n={s:.6f}  ln(n)={math.log(k):.6f}")

print("\nAlternating harmonic sum (-1)^n/n = -ln 2 =", -math.log(2))
for k, s in partial_sums(lambda n: (-1)**n / n, 100_000):
    if k in checkpoints:
        print(f"  n={k:>7d}  s_n={s:.8f}  err={abs(s + math.log(2)):.2e}")
```

## 3. Taylor expansion of $e^x$

Taylor's theorem: for smooth $f$,
$$
  f(x) = \sum_{k=0}^{N} \frac{f^{(k)}(x_0)}{k!} (x - x_0)^k + R_N(x).
$$
For $f(x) = e^x$ centred at $0$ we get $e^x = \sum_{k=0}^{\infty} x^k / k!$,
which converges for every $x \in \mathbb{R}$ (ratio test: $|x|/(n+1) \to 0$).


```python
import math

def taylor_exp(x, N):
    s = 0.0
    term = 1.0  # x^0 / 0!
    for k in range(N + 1):
        s += term
        term *= x / (k + 1)
    return s

target = math.e
for N in (0, 1, 2, 3, 5, 10, 15, 20):
    approx = taylor_exp(1.0, N)
    print(f"N={N:>3d}  sum={approx:.12f}  err={abs(approx - target):.2e}")
print(f"target e = {target:.12f}")
```
