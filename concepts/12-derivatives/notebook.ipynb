# Derivatives (Univariate)

Concept **12-derivatives** of the math-foundations map.

The derivative captures the **slope of the local linear approximation**
to a function at a point. It is the limit of secant-line slopes:

$$
f'(c) \;=\; \lim_{h \to 0} \frac{f(c+h) - f(c)}{h}.
$$

In this notebook we:

1. Recall the formal definition.
2. Verify numerically that the difference quotient for $f(x) = x^2$ at $c = 2$ converges to $4$.
3. Verify the chain rule on $g(x) = \sin(x^2)$ at $x = 1$.
4. Leave you with three exercises.


## 1. Definition

Let $f$ be defined on an open interval containing $c$. We say $f$ is
**differentiable at $c$** when

$$
f'(c) \;=\; \lim_{h \to 0} \frac{f(c+h) - f(c)}{h}
$$

exists and is finite. Equivalently, $f$ admits a local linear
approximation

$$
f(c+h) \;=\; f(c) + f'(c)\, h + o(h) \qquad (h \to 0).
$$

**Differentiable $\Rightarrow$ continuous** (proved in `lesson.tex`):
write
$f(x) - f(c) = \frac{f(x) - f(c)}{x - c} \cdot (x - c)$
and take limits.


## 2. Numerical check: $f(x) = x^2$ at $c = 2$

Analytically $f'(2) = 4$. We compute the forward difference quotient
for $h = 10^{-1}, 10^{-2}, \ldots, 10^{-10}$ and watch convergence.
At very small $h$, floating-point cancellation re-introduces error.


```python
import math

def diff_quotient(f, c, h):
    return (f(c + h) - f(c)) / h

f = lambda x: x * x
c = 2.0
true_val = 4.0

print(f"{'h':>10}  {'approx':>22}  {'|error|':>12}")
for k in range(1, 11):
    h = 10.0 ** (-k)
    approx = diff_quotient(f, c, h)
    print(f"{h:>10.1e}  {approx:>22.15f}  {abs(approx - true_val):>12.2e}")
```

## 3. Chain rule

If $g$ is differentiable at $c$ and $f$ is differentiable at $g(c)$,
then

$$
(f \circ g)'(c) \;=\; f'(g(c)) \cdot g'(c).
$$

Take $g(x) = \sin(x^2)$. Then $g'(x) = \cos(x^2) \cdot 2x$, so at
$x = 1$ we should get $g'(1) = 2 \cos 1$. We verify with a central
difference, which is more accurate than the one-sided version.


```python
import math

g = lambda x: math.sin(x * x)
x0 = 1.0
analytic = 2.0 * x0 * math.cos(x0 * x0)

h = 1e-6
central = (g(x0 + h) - g(x0 - h)) / (2 * h)

print(f"Analytic  g'(1) = 2 cos(1) = {analytic:.15f}")
print(f"Central diff (h={h:g})    = {central:.15f}")
print(f"|error|                    = {abs(central - analytic):.2e}")

assert abs(central - analytic) < 1e-8
print("\nChain rule verified numerically.")
```
