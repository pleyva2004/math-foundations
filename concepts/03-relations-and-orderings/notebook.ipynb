## Definitions

Let $R \subseteq A \times A$ be a relation on a set $A$.

- **Reflexive**: $\forall a,\ a \mathrel{R} a$.
- **Symmetric**: $a \mathrel{R} b \Rightarrow b \mathrel{R} a$.
- **Transitive**: $a \mathrel{R} b \wedge b \mathrel{R} c \Rightarrow a \mathrel{R} c$.
- **Antisymmetric**: $a \mathrel{R} b \wedge b \mathrel{R} a \Rightarrow a = b$.

An **equivalence relation** is reflexive + symmetric + transitive. A **partial order** is reflexive + antisymmetric + transitive. A **total order** is a partial order in which every pair is comparable.


```python
from itertools import product

def is_reflexive(R, A):    return all((a, a) in R for a in A)
def is_symmetric(R):       return all((b, a) in R for (a, b) in R)
def is_transitive(R):
    P = set(R)
    return all((a, d) in P for (a, b) in P for (c, d) in P if b == c)
def is_antisymmetric(R):
    return all(not (a != b and (b, a) in R) for (a, b) in R)

# Equivalence relation: x = y (mod 5) on {0,...,19}
A = list(range(20))
mod5 = {(x, y) for x in A for y in A if (x - y) % 5 == 0}
print("reflexive   :", is_reflexive(mod5, A))
print("symmetric   :", is_symmetric(mod5))
print("transitive  :", is_transitive(mod5))
print("size of R   :", len(mod5))
```

## Equivalence classes form a partition

Each integer $a \in \{0,\dots,19\}$ has class $[a] = \{ x : x \equiv a \pmod 5 \}$. Distinct classes are disjoint, and together they cover the whole set -- a *partition*.


```python
def equivalence_classes(R, A):
    classes, seen = [], set()
    for a in A:
        if a in seen:
            continue
        cls = frozenset(b for b in A if (a, b) in R)
        classes.append(cls); seen |= cls
    return classes

classes = equivalence_classes(mod5, A)
print("number of classes:", len(classes))
for c in sorted(classes, key=min):
    print(" ", sorted(c))

# verify partition
union = set().union(*classes)
disjoint = all((not (B & C)) for i, B in enumerate(classes) for C in classes[i+1:])
print("covers A?       :", union == set(A))
print("pairwise disjoint:", disjoint)
```

## A partial order: divisibility on $\{1,\dots,12\}$

Define $a \le b$ iff $a \mid b$. Reflexive ($a \mid a$), antisymmetric (for positive integers), transitive. But $2 \nmid 3$ and $3 \nmid 2$, so the order is **not total**.


```python
B = list(range(1, 13))
div = {(a, b) for a in B for b in B if b % a == 0}
print("reflexive     :", is_reflexive(div, B))
print("antisymmetric :", is_antisymmetric(div))
print("transitive    :", is_transitive(div))
print("symmetric     :", is_symmetric(div), "(expected False)")
incomp = [(a, b) for a in B for b in B if a < b
          and (a, b) not in div and (b, a) not in div]
print("first 5 incomparable pairs:", incomp[:5])
```
