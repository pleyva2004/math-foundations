We numerically illustrate:
1. Two independent fair coin flips — empirical $P(A\cap B) \to P(A) P(B)$.
2. The pairwise-but-not-mutually-independent triple on $\{1,2,3,4\}$.
3. Independence of complements.
4. Conditional independence example.


## 1. Two independent coin flips

Let $A = \{X_1 = H\}$ and $B = \{X_2 = H\}$ for two fair flips. We expect $P(A \cap B) = 0.25$.


```python
import random
from itertools import product

def coin_demo(n=200_000, seed=0):
    rng = random.Random(seed)
    nA = nB = nAB = 0
    for _ in range(n):
        a = rng.random() < 0.5
        b = rng.random() < 0.5
        nA += a; nB += b; nAB += a and b
    return nA/n, nB/n, nAB/n

pA, pB, pAB = coin_demo()
print(f'P(A)={pA:.4f}  P(B)={pB:.4f}  P(A and B)={pAB:.4f}  P(A)P(B)={pA*pB:.4f}')
```

## 2. Pairwise but not mutually independent

On $\Omega = \{1,2,3,4\}$ uniform, take $A=\{1,2\}, B=\{1,3\}, C=\{1,4\}$.
Each pair is independent but the triple is not.


```python
omega = {1,2,3,4}
A, B, C = {1,2}, {1,3}, {1,4}
P = lambda S: len(S)/len(omega)
print('P(A)P(B)      =', P(A)*P(B),     '  P(A∩B)       =', P(A&B))
print('P(A)P(C)      =', P(A)*P(C),     '  P(A∩C)       =', P(A&C))
print('P(B)P(C)      =', P(B)*P(C),     '  P(B∩C)       =', P(B&C))
print('P(A)P(B)P(C)  =', P(A)*P(B)*P(C),'  P(A∩B∩C)     =', P(A&B&C))
```

## 3. Independence of complements

If $A \perp B$, then $A^c \perp B$. Verify numerically.


```python
# Reuse the coin example. A^c = {X1=T}.
rng = random.Random(1)
n = 200_000
nAc = nB = nAcB = 0
for _ in range(n):
    a = rng.random() < 0.5
    b = rng.random() < 0.5
    nAc += (not a); nB += b; nAcB += (not a) and b
print(f'P(A^c)={nAc/n:.4f}  P(B)={nB/n:.4f}  P(A^c and B)={nAcB/n:.4f}  P(A^c)P(B)={nAc*nB/n/n:.4f}')
```

## 4. Conditional independence

Two unconditionally independent fair coins remain independent conditional on, say, $C = \{X_1 = H\}$. We verify $P(A \cap B \mid C) = P(A \mid C) P(B \mid C)$.


```python
rng = random.Random(2)
n = 200_000
nC = nAC = nBC = nABC = 0
for _ in range(n):
    a = rng.random() < 0.5  # X1=H
    b = rng.random() < 0.5  # X2=H
    c = a                    # condition on X1=H
    if c:
        nC += 1
        if a: nAC += 1
        if b: nBC += 1
        if a and b: nABC += 1
if nC:
    pA_C = nAC/nC; pB_C = nBC/nC; pAB_C = nABC/nC
    print(f'P(A|C)={pA_C:.4f}  P(B|C)={pB_C:.4f}  P(A∩B|C)={pAB_C:.4f}  P(A|C)P(B|C)={pA_C*pB_C:.4f}')
```

## Takeaways

- Independence is a multiplicative product rule, not a vague notion of unrelatedness.
- Pairwise independence is strictly weaker than mutual independence.
- IID assumptions in ML papers (e.g. independent rollouts in GRPO) rely on this exact definition.
